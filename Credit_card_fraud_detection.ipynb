```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

### read

```python
df = pd.read_csv('credit_card.csv')
```

```python
df
```

### describe

```python
df.isnull().sum()
```

```python
df.Class.value_counts()
```

unbalance dataset, very little information about fraud operations

### split dataset into fraud and normal

```python
normal = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
normal.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
df.groupby('Class').mean()
```

### create a sample of normal operations whose length is equal to the length of fraudulent ones

```python
normal_sample = normal.sample(n=len(fraud))
```

```python
normal_sample
```

```python
new_df = pd.concat([normal_sample,fraud])
new_df = new_df.reset_index(drop=True)
```

```python
new_df
```

### Splitting into features and targets

```python
X = new_df.drop(columns='Class', axis=1)
Y = new_df['Class']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

### 1st Model: Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

accuracy on training data:

```python
prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

accuracy on test data:

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
